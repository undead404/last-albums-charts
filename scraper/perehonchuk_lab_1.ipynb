```octave
T = [30 31 30.5];
dt = 0.1;
n = numel(T)
t = (0:1:n - 1) * dt;
```

```octave
t
T
```

```octave
n = numel(t)
```

```octave
min_T = min(T)
mean_T = mean(T)
std_T = std(T)
```

```octave
plot(t, T - mean_T, '-ro')
title('Temperature')
xlabel('time, s')
ylabel('temperature, ^oC')
%xlim([0 0.15])
grid on
```
